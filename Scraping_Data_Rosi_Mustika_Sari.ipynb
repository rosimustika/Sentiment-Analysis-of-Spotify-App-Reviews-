```python
from google_play_scraper import reviews, Sort
import pandas as pd
import time
import re
```

```python


all_reviews = []
count_per_batch = 1000
max_data = 20000
token = None

print("Mulai scraping...")

while len(all_reviews) < max_data:
    result, token = reviews(
        'com.spotify.music',
        lang='id',
        country='id',
        sort=Sort.NEWEST,
        count=count_per_batch,
        continuation_token=token
    )
    all_reviews.extend(result)
    print(f"✅ Total terkumpul: {len(all_reviews)}")

    time.sleep(1)


    if token is None:
        break

print(f"\nScraping selesai! Total data terkumpul: {len(all_reviews)}")

# Konversi ke DataFrame
df = pd.DataFrame(all_reviews)
df = df[['userName', 'content', 'at']].copy()
df['at'] = pd.to_datetime(df['at'])

df.to_csv('spotify_reviews.csv', index=False)

print("\nFile tersimpan sebagai spotify_reviews.csv")
```
